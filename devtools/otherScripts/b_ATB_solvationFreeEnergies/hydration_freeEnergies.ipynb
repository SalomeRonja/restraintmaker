```python
import matplotlib
from matplotlib import pyplot as plt

from scipy import stats
import numpy as np
```

```python
from test_set_ATB import solvation_free_energy_kJ
solvation_free_energy_kJ.head()
```

```python
lin_reg = stats.linregress(solvation_free_energy_kJ.exp, solvation_free_energy_kJ.exp)
kcal_to_kJ = 4.18
min_x, max_x = (-50, 5)
x_opt = np.arange(min_x, max_x)
y_opt = lin_reg.slope*x_opt+lin_reg.intercept
y_opt_kcal = lin_reg.slope*x_opt+lin_reg.intercept+kcal_to_kJ
y_opt_mkcal = lin_reg.slope*x_opt+lin_reg.intercept-kcal_to_kJ

plt.plot(x_opt, y_opt, c="k")
plt.plot(x_opt, y_opt_kcal, linestyle="--", c="grey")
plt.plot(x_opt, y_opt_mkcal, linestyle="--", c="grey")

plt.ylim([min_x, max_x])
plt.xlim([min_x, max_x])

plt.xlabel("experiment [kJ]")
plt.ylabel("TI [kJ]")

#DATA:
plt.scatter(solvation_free_energy_kJ.exp, solvation_free_energy_kJ.TI, zorder=-1)
```
