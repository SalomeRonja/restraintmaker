```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## 3 Restraints

```python
data_3disres = pd.read_csv("./disres_n3/tmp_algorithm_ana.tsv", sep="\t")
```

```python
data_3disres.columns
```

```python
dists = list(map(lambda x: list(map(float, x.replace("[", "").replace("]", "").split(","))), data_3disres.myBruteF_all_d))
dists = np.array(dists)

lables = list(data_3disres[data_3disres.columns[0]].values)

averages = np.array([np.mean(x) for x in dists])
ndiss = [np.array(d)/avd for avd, d in zip(averages, dists)] #norm with toatl V
plt.violinplot(dists)
#plt.boxplot(dists)
plt.scatter(x=list(range(1, 1+len(dists))) , y=data_3disres.greedy_shortest_distance, alpha=0.4, zorder=100, label="greedy_shortest")
plt.scatter(x=list(range(1, 1+len(dists))) , y=data_3disres.greedy_cog_distance, zorder=100, alpha=0.4, label="greedy_cog")
plt.scatter(x=list(range(1, 1+len(dists))) , y=data_3disres.greedy_biasedavg_distance, zorder=100, alpha=0.4, label="greedy_biasAvg")
plt.scatter(x=list(range(1, 1+len(dists))) , y=data_3disres.greedy_prim_distance, zorder=100, alpha=0.4, label="greedy_prim")

plt.legend()
plt.xticks(list(range(len(lables))), lables, rotation=45)
plt.ylabel("total restraint distance[nm]")
#plt.yticks(np.linspace(10,20, 5), np.round(np.linspace(10,20, 5)/10,2))
plt.savefig("distance_analysis.png")
;
```

```python
plt.bar(x=data_3disres.index, height=data_3disres.total_BruteF_t)
plt.ylabel("t [s]")
plt.xticks(list(range(-1,-1+len(lables))), lables, rotation=45)
plt.ylim([0,30])
```
