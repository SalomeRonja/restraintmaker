```python
dists = list(map(lambda x: list(map(float, x.replace("[", "").replace("]", "").split(","))), volumes.bruteF_d))
dists = np.array(dists)

lables = list(volumes[volumes.columns[0]].values)

averages = np.array([np.mean(x) for x in dists])
ndiss = [np.array(d)/avd for avd, d in zip(averages, dists)] #norm with toatl V
plt.violinplot(dists)
plt.boxplot(dists)
plt.scatter(x=list(range(1, 1+len(dists))) , y=volumes.greed_d, zorder=100, label="greedy")
plt.xticks(list(range(len(lables))), lables, rotation=45)
plt.ylabel("total restraint distance[nm]")
plt.yticks(np.linspace(10,20, 5), np.round(np.linspace(10,20, 5)/10,2))
plt.savefig("distance_analysis.png")
;
```

```python
dists = list(map(lambda x: list(map(float, x.replace("[", "").replace("]", "").split(","))), volumes.myBruteF_dist))
dists = np.array(dists)

lables = list(volumes[volumes.columns[0]].values)

averages = volumes.mol #np.array([np.mean(x) for x in dists])
ndiss = [np.array(d)/avd for avd, d in zip(averages, dists)] #norm with toatl V
plt.violinplot(ndiss)
plt.boxplot(ndiss)
plt.scatter(x=list(range(1, 1+len(dists))) , y=volumes.greed/averages)
plt.xticks(list(range(len(lables))), lables, rotation=45)
plt.ylabel("convexHull Volumes [A^3]")
#plt.yticks(np.linspace(0,20, 5), np.round(np.linspace(0,20, 5),2))
plt.savefig("convexH_volume_analysis.png")

;
```

```python
delta = volumes.brutF - volumes.greed
norm_delta =  delta / volumes.mol
plt.bar(x=volumes.index, height=norm_delta)
lables = list(volumes[volumes.columns[0]].values)
plt.xticks(list(range(-1,-1+len(lables))), lables, rotation=45)
plt.ylim([0,0.75])
plt.ylabel("delta fraction of found convex hull volume")
;
```
